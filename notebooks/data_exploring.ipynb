maybe use them for analysis instead of removing?

to transform and standardise names, remove all the feats (usually in brackets)

```python
#df['track_name'] = df['track_name'].str.replace(r'\s*\(.*?\)|\s*\[.*?\]', '', regex=True)

def remove_features_from_name(df):
    df['track_name'] = df['track_name'].str.replace(r'\s*\(.*?\)|\s*\[.*?\]', '', regex=True)
    return df
    
```

```python
data = remove_features_from_name(data)
```

```python
data.head(10)
```

Now lets add a features column using the artist_name

```python
def add_features_column(df):
    df['features'] = df['artist_name'].apply(lambda x: x.split(', ')[1:] if ',' in x else [])
    return df
```

```python
add_features_column(data)
```

now I want to remove features from the artist_name column

```python
def remove_features_from_artist_name(df):
    df['artist_name'] = df['artist_name'].apply(lambda x: x.split(',')[0])
    return df
```

```python
data = remove_features_from_artist_name(data)
```
